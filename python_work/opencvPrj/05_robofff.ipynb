```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")  # nano 모델

model.train(
    data="../fff/cat_dog_tiger/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16
)
```

```python
## 이미지 추론

from ultralytics import YOLO

model = YOLO("runs/detect/train/weights/best.pt")

results = model("test.jpg")

for r in results:
    for box in r.boxes:
        cls = int(box.cls[0])

        print(
            model.names[cls],
            float(box.conf[0])
        )
```

```python
from ultralytics import YOLO

model = YOLO("best.pt")

results = model.predict(
    source="test.jpg",
    save=True,
    conf=0.25
)
```

```python
from ultralytics import YOLO
import cv2

model = YOLO("best.pt")

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    results = model(frame)

    annotated = results[0].plot()

    cv2.imshow("YOLO11", annotated)

    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
