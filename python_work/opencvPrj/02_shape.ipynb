### 도형그리기

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

## 선그리기
## 검은색 배경 이미지 생성 200x500
space = np.zeros((200,500),dtype=np.uint8)
print(space)

#cv2.line : 직선그리기
#space : 그리는 위치
# 50,50 : 시작점
# 400,150 : 끝점
# 255 : 선색
# 3 : 선 두께
# 1 : 선타입
space = cv2.line(space,(50,50),(400,150),255,3,1)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
space = np.zeros((200,500),dtype=np.uint8)

##cv2.circle : 원그리기
# space : 그리기 위치
# (250,100): 원의 중심
# 50 : 반지름
# 255 : 색상
# 4 : 두께
space = cv2.circle(space,(250,100),50,255,4)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
## BGR 3가지 색이므로  (200,500,3)
space = np.zeros((200,500,3),dtype=np.uint8)

##cv2.circle : 원그리기
# space : 그리기 위치
# (250,100): 원의 중심
# 50 : 반지름
# (255,0,0) : 색상(BGR)
# 4 : 두께
space = cv2.circle(space,(250,100),50,(255,0,0),4)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
space = np.zeros((200,500,3),dtype=np.uint8)
##cv2.rectangle 직사각형 그리기
# space : 그리기영역
# (50,30) : 시작점
# (400,80) : 끝점
# (0,0,255) : 색상
# 5 : 선두께
# 1 : 선타입
space = cv2.rectangle(space,(50,30),(400,80),(0,0,255),5,1)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
space = np.zeros((200,500,3),dtype=np.uint8)

## cv2.ellipse : 타원(호) 그리기
# space : 그리기 영역
# (50,30) : 중심점
# (200,80) : x, y 반지름
# 30 : 회전각
# 0 : 시작각 
# 90 :끝각
# (0,0,255) :색상
# 5 : 두께
space = cv2.ellipse(space,(200,100),(100,50),20,0,180,(0,0,255),5)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
space = np.zeros((200,500,3),dtype=np.uint8)

## 다각형(선) 그리기
## 각 점의 위치
sh1 = np.array([[100,10],[300,10],[250,100],[50,100]])
#cv2.polylines : 다각형(선) 그리기
# space : 그리기 영역
# [sh1] : 각 점의 좌표 (3차원)
# True : 닫힌 도형
# (0,255,0) : 색상
# 3 : 선두께
space = cv2.polylines(space,[sh1],True, (0,255,0),3)

## 다각형(면) 그리기
## 각 점의 위치
sh2 = np.array([[350,50],[230,150],[470,150]])
#cv2.fillPoly : 다각형 그리기(면채우기)
# space : 그리기 영역
# [sh2] : 각 점의 좌표 (3차원)
# (255,255,0) : 색상
space = cv2.fillPoly(space,[sh2],(255,255,0))

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
## 격자 그리기
space = np.zeros((200,500,3),dtype=np.uint8)
print(space.shape)
## (200, 500, 3)  y의 크기, x의 크기, 색상채널갯수 


## 세로선 그리기
# space = cv2.line(space,(50,0),(50,500),(0,0,255),2)
# space = cv2.line(space,(100,0),(100,500),(0,0,255),2)
for x in range(0,space.shape[1],50):
    space = cv2.line(space,(x,0),(x,500),(0,0,255),2)
    
# 가로선 그리기
# space = cv2.line(space,(0,50),(500,50),(255,0,0),2)
# space = cv2.line(space,(0,100),(500,100),(255,0,0),2)
for y in range(0,space.shape[0],50):
    space = cv2.line(space,(0,y),(500,y),(255,0,0),2)
    

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```
