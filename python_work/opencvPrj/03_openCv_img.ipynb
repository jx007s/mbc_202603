```python
import cv2
import matplotlib.pyplot as plt

## 이미지 로드 (흑백처리)
img = cv2.imread('../fff/three young man.jpg',cv2.IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#하르케스케이드 로드
## 정면 얼굴 검출용 학습데이터
face_cascade=cv2.CascadeClassifier(
    cv2.data.haarcascades+"haarcascade_frontalface_default.xml"
)

faces = face_cascade.detectMultiScale(
    img,                ##원본
    scaleFactor=1.1,    # 이미지 크기를 줄여가며 얼굴 탐색 1.1 매 단계시 10%축소
    minNeighbors=5,     # 얼굴 데이터중 5개 이상 폅치면 얼굴로 인정
    minSize=(30,30)     ## 최소 얼굴 30*30 이상만 검색
)

print(faces)
'''
[x,y,w,h]   얼굴위치 : x,y
            얼굴크기 : w,h
[[449 111 173 173]
 [721 106 170 170]
 [159 148 160 160]
 [371 376 116 116]]
'''
```
