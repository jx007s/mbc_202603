### 도형그리기

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

## 선그리기
## 검은색 배경 이미지 생성 200x500
space = np.zeros((200,500),dtype=np.uint8)
print(space)

#cv2.line : 직선그리기
#space : 그리는 위치
# 50,50 : 시작점
# 400,150 : 끝점
# 255 : 선색
# 3 : 선 두께
# 1 : 선타입
space = cv2.line(space,(50,50),(400,150),255,3,1)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
space = np.zeros((200,500),dtype=np.uint8)

##cv2.circle : 원그리기
# space : 그리기 위치
# (250,100): 원의 중심
# 50 : 반지름
# 255 : 색상
# 4 : 두께
space = cv2.circle(space,(250,100),50,255,4)

img = cv2.cvtColor(space, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.axis('off')
plt.imshow(img)
plt.show()
```
