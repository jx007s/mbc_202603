```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('../fff/like_lenna.png')  ##BGR 방식으로 읽음

## matplotlib 방식 RGB  <=  BGR 방식을 변환
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

## 이미지생성
plt.imshow(img)

## 화면에 표시
plt.show()
```

### 이미지변환

```python

img = cv2.imread('../fff/like_lenna.png')
img_s = cv2.resize(img,(100,100))       ##이미지 크기변경
cv2.imshow("img_s",img_s)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
hh = 300
ww = 300
dst = np.zeros((hh,ww), dtype=np.uint8)
print(dst)
## 원본 : 100x100  --> 300x300 이미지 깨짐현상 발생
img_g = cv2.resize(img_s,(ww,hh),dst=dst)
cv2.imshow("img_g",img_g)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../fff/like_lenna.png', cv2.IMREAD_GRAYSCALE)
img_b = cv2.resize(img,(200,200))       ##이미지 크기변경
cv2.imshow("img_b",img_b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 대칭

```python
img_f = cv2.flip(img_b,0)  ## 0 상하 대칭

img = cv2.cvtColor(img_f, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(img)
plt.show()
```

```python
img_f = cv2.flip(img_b,1)  ## 1 좌우 대칭

img = cv2.cvtColor(img_f, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(img)
plt.show()
```

### 회전

```python
print(img_b.shape) ## shape : 현재 자료형태
hh,ww = img_b.shape
##회전 행렬 생성
#(ww/2,hh/2) : 회전 중심축
# 90 : 회전 각도
# 1 : 확대축소 비율
matrix = cv2.getRotationMatrix2D((ww/2,hh/2),90,1)

##warpAffine :변환
##img_b : 원본
# matrix : 변환값
# (ww,hh) : 결과이미지 크기
res = cv2.warpAffine(img_b, matrix,(ww,hh))

img = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(img)
plt.show()
```

```python
matrix = cv2.getRotationMatrix2D((ww/2,hh/2),30,1)
## 빈공간 채우기 : borderValue=(200,200,200)  : RGB 0~255
res = cv2.warpAffine(img_b, matrix,(ww,hh), borderValue=(200,200,200))

img = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(3,3))
plt.imshow(img)
plt.show()
```
