```python
import easyocr

reader = easyocr.Reader(['ko','en']) ##  한국어 + 영어
res = reader.readtext("../fff/card.png")
```

```python
res
##[
    ## 1 단어 시작 : tuple
    ## r[0] : 단어좌표 : list
    ## r[1] : 내용
    ## r[2] : 신뢰도
# ([[np.int32(214), np.int32(52)],
#    [np.int32(242), np.int32(52)],
#    [np.int32(242), np.int32(80)],
#    [np.int32(214), np.int32(80)]],
#   '김',
#   np.float64(0.9999046348420961)),
##  단어끝
#...    
#]
```

```python
for r in res:
    print(r[1], r[2])  ## r[0] 좌표, r[1] 내용, r[2]신뢰도
```

```python
res = reader.readtext(
    "../fff/card.png",
    detail=0        ## 0: 내용만 리턴, 1: 위치, 신뢰도포함(기본값)
    )
print(res)
```
